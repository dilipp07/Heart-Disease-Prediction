```python
numerical_columns=[features for features in df.columns if df[features].dtype!='O']
numerical_columns
```

```python
categorical_columns=[features for features in df.columns if df[features].dtype=='O']
categorical_columns
```

```python
df[categorical_columns].describe() #not a imbalance dataset
```

```python
for i in numerical_columns:
    sns.distplot(df[i],kde=True)
    print('\n')
    plt.show()
```

```python
# Most Of the Feature Follows Normal Distribution
```

```python
df.corr()
```

```python
# we find That there is hardly any correlation between the features
```

```python
# Box plots
plt.figure(figsize=(10, 6))
for col in numerical_columns:
    plt.subplot(2, 2, 1)
    sns.boxplot(x='Heart Disease', y=col, data=df)
    plt.title(f'Box Plot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
for col in df.select_dtypes(include=['object']):
    plt.subplot(1, 2, 1)
    sns.countplot(data=df, x=col, hue='Heart Disease')
    plt.title(f'Count Plot of {col}')
plt.tight_layout()
plt.show()

```

```python
# Scatter plots
sns.pairplot(df, hue='Heart Disease')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
